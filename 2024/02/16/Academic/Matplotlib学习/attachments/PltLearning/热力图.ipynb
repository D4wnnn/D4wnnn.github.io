```python
import matplotlib.pyplot as plt
import numpy as np
import random

# 让图片可以显示中文
plt.rcParams['font.sans-serif'] = 'SimHei'
# 让图片可以显示负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
xNum = 9
yNum = 6
xLabel = [chr(ord('A') + i) for i in range(xNum)]
print(xLabel)
yLabel = [chr(ord('A') + i + xNum) for i in range(yNum)]
print(yLabel)
```

```python
data = [[random.randint(100, 6000) for _ in range(xNum)] for _ in range(yNum)]
print(data)
```

```python
ax1 = plt.subplot(2, 2, 1)
ax1.set_title('Blues')
ax1.imshow(data, cmap='Blues')

ax2 = plt.subplot(2, 2, 2)
ax2.set_title('cool')
ax2.imshow(data, cmap='cool')

ax3 = plt.subplot(2, 2, 3)
ax3.set_title('rainbow')
ax3.imshow(data, cmap='rainbow')

ax4 = plt.subplot(2, 2, 4)
ax4.set_title('Blues_r')
ax4.imshow(data, cmap='Blues_r')
```

```python
plt.imshow(data, cmap='Blues_r')
# 添加刻度
plt.xticks(range(xNum), xLabel)
plt.yticks(range(yNum), yLabel)
# 添加文字
for i in range(yNum):
    for j in range(xNum):
        plt.text(
            y=i,
            x=j,
            s=data[i][j],
            ha='center',
            va='center',
            fontsize=15
        )
# 添加边栏
plt.colorbar(ticks=[0, 1000, 2000, 3000, 4000, 5000], shrink=0.7, orientation='vertical')
```
